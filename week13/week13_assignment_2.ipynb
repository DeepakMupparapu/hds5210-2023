### 1. Basic Statistics

Use Python to read in your data file and show many rows and columns your data has.

```python
import pandas as pd
import gdown

# Specify the Google Drive file ID
file_id = '1ytVS3_Itkwn1Rxum5o2F5ak2-NfeBk7W'

# Construct the download link
url = f'https://drive.google.com/uc?id={file_id}'

# Specify the local file path for saving the downloaded file
output_path = 'hhcaps.csv'

# Download the file
gdown.download(url, output_path, quiet=False)

# Read the CSV file into a DataFrame
hhcaps = pd.read_csv(output_path)

# Display the first few rows of the DataFrame
print("First few rows of the DataFrame:")
display(hhcaps.head())

# Display information about the DataFrame (rows, columns, data types, memory usage, etc.)
hhcaps.info()
```

### 2. Data Diversity

Use Python to identify how many unique values each column in your data has.  Use Python to identify which column has the greatest number of distinct values (aka the most diversity)?  If your data has a unique identifier (e.g. a person or visit or record ID) then find the column with the next most number of unique values.

```python
# Identify the number of unique values in each column
unique_counts = hhcaps.nunique()

# Display the number of unique values in each column
print("Number of unique values in each column:")
print(unique_counts)

# Find the column with the greatest number of distinct values
max_unique_column = unique_counts.idxmax()
max_unique_count = unique_counts.max()

print("\nColumn with the most unique values:")
print(f"Column: {max_unique_column}")
print(f"Number of unique values: {max_unique_count}")

# If there's a unique identifier column, find the column with the next most number of unique values
unique_counts.drop(max_unique_column, inplace=True)
second_max_unique_column = unique_counts.idxmax()
second_max_unique_count = unique_counts.max()

print("\nColumn with the second most unique values:")
print(f"Column: {second_max_unique_column}")
print(f"Number of unique values: {second_max_unique_count}")
```

### 3. Data Redundancy

Use Python to identify any columns in your data where the value for every row is the same.  For example, if you had a gender column and all the rows said 'M', then you would want to write the code that helped you determine that.

```python
# Identify columns where the value for every row is the same
constant_columns = [col for col in hhcaps.columns if hhcaps[col].nunique() == 1]

# Display the list of columns with constant values
print("Columns where the value for every row is the same:")
print(constant_columns)
```

### 4. Range of Values

Your data almost certainly has some numeric columns that can be summed or averaged.  Create a histogram that shows the distribution of values for this column.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Assuming 'bmi' and 'age' are the column names in your DataFrame
# Replace them with the actual column names if they are different
columns_of_interest = ['bmi', 'age']

# Create histograms for each specified column
for column in columns_of_interest:
    plt.figure(figsize=(8, 5))
    plt.hist(hhcaps[column], bins=20, color='blue', edgecolor='black')
    plt.title(f'Distribution of {column}')
    plt.xlabel(column)
    plt.ylabel('Frequency')
    plt.grid(True)
    plt.show()
```
